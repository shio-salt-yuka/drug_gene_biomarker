```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
from scipy.stats import variation
from sklearn.cluster import KMeans


a_dataframe = pd.read_csv("HCI.gr50.scores.txt",sep="\t")
#sorting textfile
df = a_dataframe[["HCIid","drug", "GR_AOC"]]
df2 = df[df.HCIid != "Control"]
df3 = df2[df2.HCIid != "HCI-011.E2"]
df4 = df3[df3.HCIid != "HCI-017.E2"]
df = df4
df = df.pivot(index='HCIid',
        columns='drug',
        values='GR_AOC')
# print(df)


drug_dict = {};
#creating dataframe with calculated data......
for x in range(len(df.columns)):
    #Kmeans, output: 2 centroid points
    drugName = df.columns[x]
    kmeans = KMeans(n_clusters = 2,random_state=0).fit(df[[df.columns[x]]])
    #print(sum(kmeans.labels_))
    nDiff = abs((len(kmeans.labels_)- sum(kmeans.labels_)) - sum(kmeans.labels_))
#smaller the nDiff, closer the sizes are between the two groups
   # print(nDiff)
    clusters = kmeans.cluster_centers_
    big = max(clusters)
    big = big[0]
    small = min(clusters)
    small = small[0]
    diff = big - small
    mean = df[drugName].mean()
    cv = variation(df[drugName])
    drug_dict[df.columns[x]] = mean, cv, big, small, diff, nDiff
    
data = pd.DataFrame.from_dict(drug_dict, orient='index')
data = data.rename(columns={0: "Mean", 1: "CV", 2: "Big", 3: "Small", 4: "Difference", 5:"nDiff"})
# print(data)
#number of samples in each clusters
#cv, more diverse/spread 
#bigger big values, more effective it was in small dose

# a_list = list(data.index)
# for i in range(len(a_list)):
#     a_list.index(i)[0:3]
# print(a_list)
#abb = abbreviate(a_list, minlength = minlength) 
#print(abb)
#abreviate names on the plot

Mean = data["Mean"]
CoeV = data["CV"]
Big = data["Big"]
Small = data["Small"]
Difference = data["Difference"]
nDiff = data["nDiff"]

labels = ["5-f","abe", "abi", "api","asl", "azd8", "azd1", "azd5", "azd6", "azd7", "bir","bkm", "car", "ced",
         "cer", "cob", "dms", "doc", "dox", "epi", "epz", "eve", "fk8", "ful", "gem", "h-t", "ibr", "lap", "lar"
         "ly3","nav", "nir", "ola", "pac", "pal", "pon", "rap", "ro4", "rom", "sap", "sn", "tas", "tuc", "vem",
          "vism","vist"]
#print(data)
#Mean
# fig, ax = plt.subplots()
# ax.scatter(nDiff, Big)
  
# for i, drug in enumerate(labels):
#     ax.annotate(drug, (nDiff[i], Big[i]))

#t-test or classification tests(LDA, Logistic regression, linear) on genes.
#kegg, wiki gene etc

df = pd.read_csv("normalized.DESEQ.RSEM.txt",sep="\t")
print(df)
# df = a_dataframe[["AM0245_pdx_PX_XXX_RNA_15408X1"]]
index = df.index
pev_index = index[df["pevonedistat"]]

print(pev_index)
```
